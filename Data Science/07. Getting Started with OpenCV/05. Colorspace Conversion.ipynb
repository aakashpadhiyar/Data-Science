```python
import cv2
import matplotlib.pyplot as plt
```

```python
impath = '/home/aakash/Downloads/data_Science/standard_test_images/peppers_color.tif'
img = cv2.imread(impath, 0)
```

```python
plt.imshow(img)
plt.title('Greayscle Image - Default Colourmap')
plt.show()
```

```python
plt.imshow(img, cmap = 'gray')
plt.title('Greayscale Image - Greay Colourmap')
plt.show()
```

```python
imgpath = '/home/aakash/Downloads/data_Science/standard_test_images/peppers_color.tif'
img = cv2.imread(impath, 1)
plt.imshow(img)
plt.title('Colour Image RGB')
plt.show()
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.title('Colour Image BGR')
plt.show()
```
