# Logical Operations

```python
img1 = plt.imread('/home/aakash/Downloads/data_Science/standard_test_images/lena_color_512.tif')
img2 = plt.imread('/home/aakash/Downloads/data_Science/standard_test_images/mandril_color.tif')

```

```python
plt.imshow(np.bitwise_and(img1, img2))
plt.show()
```

```python
plt.imshow(np.bitwise_or(img1, img2))
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.imshow(np.bitwise_xor(img1, img2))
plt.subplot(1, 2, 2)
plt.imshow(np.bitwise_xor(img2, img1))
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap = 'gray')
plt.subplot(1, 2, 2)
plt.imshow(np.bitwise_not(img2), cmap = 'gray')
plt.show()
```
