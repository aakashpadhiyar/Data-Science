# Matplotlib.pyplot.plt()

## Single Line Plots in details

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 4, 5, 2, 3, 6]

plt.plot(x)
plt.show()
```

```python
import numpy as np

x = np.arange(10)

plt.plot(x)
plt.show()
```

```python
plt.plot(x, [y**2 for y in x])
plt.show()
```

```python
plt.plot(x, x**2)
plt.show()
```
